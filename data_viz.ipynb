```python
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
```

```python
gdf= gpd.read_file('/Users/sharaishuro/Documents/Shemaps/8%_project/gender_data/classified.shp')
gdf
```

```python
# Filter roads named after women
fem_road = gdf[gdf['gender'] == 'female']

# Create figure and axis with a transparent background
fig, ax = plt.subplots(figsize=(10, 8), facecolor='none')  # Figure background is transparent

# Plot roads with a bright color for contrast
fem_road.plot(ax=ax, color='darkorange', linewidth=1.5, alpha=0.9)

# Set axis background to transparent
ax.set_facecolor('none')  # Fully transparent background

# Remove axis ticks & frame for a clean floating look
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Add a stylish title
plt.title("Roads Named After Women", fontsize=14, fontweight='bold', color='white', pad=15)

 # Make legend text match road color

# Save the figure with transparency
plt.savefig("femroads_transparent.png", dpi=300, transparent=True)  # Saves with a transparent background

# Show the plot
plt.show()
```
